#Importing the dependencies

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
voc_size = 10000
```

#One Hot Representation

```python
oh_rep = [one_hot(words , voc_size) for words in sent]
```

```python
print(oh_rep)
```

We can see that the first and second sentences are almost same except milk and juice. So, in one hot representation also we can see that the first three index are same, only change is  with the last word.

#Word Embedding representation


```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
import numpy as np
```

**Padding the sentences:**

To make all the sentences in the input of the same size.

```python
sentence_length = 8
padded_sentences = pad_sequences(oh_rep, padding = 'pre', maxlen = sentence_length)
```

```python
print(padded_sentences)
```

```python
dimenssions = 10
```

#Creating a neural network

```python
model = Sequential()
```

```python
model.add(Embedding(voc_size, dimenssions, input_length = sentence_length))
```

```python
model.compile(optimizer = 'adam', loss = 'MSE')
```

```python
model.summary()
```

#Prediction

```python
print(model.predict(padded_sentences))
```

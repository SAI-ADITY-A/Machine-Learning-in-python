#Data Standardization

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(dataset)
```

```python
# loading the data to a pandas dataframe
df = pd.DataFrame(dataset.data, columns = dataset.feature_names)
```

```python
df.head()
```

```python
df.shape
```

```python
# the dataframe, df is all of the features it doesn't contain the target column (i.e whether the person has cancer or not)
# we need not standardize the target data, but we have to standardize the features dataframe(df)

x = df
y = dataset.target
```

```python
print(x)
```

```python
print(y)
```

```python
#splitting the data into training data and test data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 3)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
#Standardize the data
# first check the standard deviation of the whole data, if already somwhat standard it must be around 1
print(dataset.data.std())
```

#Standardization: It standardizes but doesn't change the nature of the data..

The score of a smaple is given as:

z = (x-u)/s



*   z = Standard value of sample
*   x = sample data
*   u = mean of the samples
*   s = standard deviation of the samples





```python
scaler = StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train_std = scaler.transform(x_train)
```

```python
print(x_train_std)
```

```python
# Now all the values are mostly similar, to standardize the remaining sets, we need not fit the data again
# We haeve to standardize all those w.r.t fit of x_train itself.
x_test_std = scaler.transform(x_test)
```

```python
print(x_test_std)
```

```python
# After standardization the std deviation of the data must be around 1
print(x_train_std.std())
```

```python
print(x_test_std.std())
```

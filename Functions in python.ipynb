#Functions
It is a block of code which can be reused throughout the program

**Factorial of a number:**

```python
num = int(input("enter a number to find out the factorial: "))
fact = 1
if num == 0:
  print("The factorial is 1")
else:
  for i in range(1,num+1):
    fact *= i
  print("The factorial is",fact)
```

**Factorial function:**

```python
def fact(num):
  factorial = 1
  if num == 0:
     return factorial
  else:
    for i in range(1,num+1):
      factorial *= i
    return factorial
```

```python
print(fact(5))
```

```python
print(fact(4))
```

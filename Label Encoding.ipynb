# Label Encoding:

Converting labels in to numeric form

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

#Label Encoding of Breast cancer data

```python
cancer_ds = pd.read_csv('/content/cancer_data.csv')
```

```python
cancer_ds.head()
```

```python
# here diagnosis column is the one that has either M or B, it has to label encoded
# finding the count of differnet labels
cancer_ds['diagnosis'].value_counts()
```

```python
# load the encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_ds.diagnosis)
```

```python
# appending the labelled column to our dataframe
cancer_ds['target'] = labels
```

```python
cancer_ds.head()
```

```python
# M ==> 1 and B ==> 0 encoded
```

# Label encoding of iris data


```python
iris_ds = pd.read_csv('/content/iris_data.csv')
```

```python
iris_ds.head()
```

```python
# here there will be three labels
iris_ds['Species'].value_counts()
```

```python
encode = LabelEncoder()
```

```python
label = encode.fit_transform(iris_ds.Species)
```

```python
# appending this labelled column to the dataframe
iris_ds['target'] = label
```

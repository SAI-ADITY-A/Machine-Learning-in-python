# **LISTS - should be stored in square brackets, can store multiple data** **points**

```python
lis = [1,2,3,4,5,6]
print(lis)
type(lis)
```

```python
ls = ["üòÇ", 1, "hi", 6.3, True, 2>3, 1 + 3j]
print(ls)
```

```python
ls.append("üòä") # adds the new data at the end of the list
print(ls[7])
```

```python
ls.append(6.3) # list allows to store duplicate values unlike sets
print(ls)
print(len(ls)) # size of the list
```

```python
print(ls)
del ls[3] # deleting an element
print(ls)
```

```python
ls1 = [1,2,3,4]
ls2 = [5,6,7,8]
ls3 = ls1 + ls2
print(ls3)
```

#**TUPLES: Immutable, they should be stored in rounded brackets (), it also allows to store multiple datatypes**

```python
tup = (1,2,3,4)
print(tup)
type(tup)
```

```python
myTup = ("Hi", 1, 6.9, "üòÅ", 1==2)
print(myTup)
```

```python
myList = [1, 2, 3, 4, 5, 6, 7, 8]
newTup = tuple(myList)
print(newTup)
type(newTup)
```

#**SET: Only stores unique elements, and should be stored using curly braces {}**



```python
mySet = {1,2,4,5}
print(mySet)
type(mySet)
```

```python
newList = [1,1,1,2,3,4,5,6]
newSet = set(newList)
print(newSet)
```

#**DICTIONARY: contains a key value pair, stored in curly brackets**



```python
dict = {'name': 'Aditya',
        'age' : 19,
        'roll' : 36,
        'country' : 'India'}
print(dict)
type(dict)
```

```python
# we can call the value using the key in the dictionary
print(dict['name'])
print(dict['country'])
```

```python
# doesn't allow duplicate values
dict1 = {'name': 'Aditya',
        'age' : 19,
        'roll' : 36,
        'country' : 'India',
        'name': 'Aditya'}
print(dict1)
```

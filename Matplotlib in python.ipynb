#**Matplotlib library:**



*   Useful for making plots
*   helps in visualization



```python
import matplotlib.pyplot as plt
```

```python
#import numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

#**Plotting the data**

```python
# sine wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding labes to x and y axis
plt.plot(x,y)
plt.xlabel('angle --->')
plt.ylabel('sine value--->')
plt.title('sine wave')
plt.show()
```

```python
# two plots in same graph
a = np.linspace(-5,5,50)
plt.plot(a,np.sin(a),'g-')
plt.plot(a,np.cos(a),'r--')
plt.xlabel('angle -->')
plt.ylabel('sine and cosine -->')
```

#**Parabola**

```python
# plotting a parabola
x = np.linspace(-10,10,50)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
```

```python
plt.plot(x,y,'g.')
```

#**Bar plots**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'Spanish', 'Japanese', 'Latin', 'German']
people = [100,56,43,50,75]
ax.bar(languages,people)
plt.xlabel('Languages -->')
plt.ylabel('No. of people -->')
plt.show()
```

#**Pie chart**

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English', 'Spanish', 'Japanese', 'Latin', 'German']
people = [100,56,43,50,75]
ax.pie(people, labels = languages, autopct = '%.2f%%') # ax.pie(y data, labels = x data, autopct = no. of decimal points u want)
plt.show()
```

#**Scatter plot**

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color = 'g')
ax.scatter(x,z,color = 'r')
plt.show()
```

#**3D scatter plot**

```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap = 'Blues')
plt.show()
```

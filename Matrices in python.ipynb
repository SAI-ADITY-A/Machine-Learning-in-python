# Matrices and working in python

```python
import numpy as np
```

1) Creating a matrix in numpy

```python
a = np.array([[1,2,3],[4,5,6]])
print(a)
a.shape
```

2) creating a random value matrix

```python
random_mat = np.random.rand(4,6)
```

```python
print(random_mat)
```

3) Creating a matrix with random integers:

```python
r = np.random.randint(100, size = (4,5))
print(r)
```

4) with all ones and zeros

```python
o = np.ones((3,5),dtype = 'int')
print(o)
```

```python
z = np.zeros((4,4), dtype = 'int')
print(z)
```

5) Identity matrix:

```python
i = np.eye(3,3)
print(i)
```

6) Transpose of a matrix:

```python
transpose_of_r = np.transpose(r)
print(transpose_of_r)
```

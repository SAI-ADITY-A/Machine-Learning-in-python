# Basic Operations on matrices in python

```python
import numpy as np
```

1) Matrix addition:

```python
a = np.random.randint(100, size = (3,3))
b = np.random.randint(50, size = (3,3))
```

```python
print(a)
print(b)
```

```python
# sum = a + b
sum = np.add(a,b)
print(sum)
```

2) Matrix Subtraction:

```python
a = np.random.randint(100, size = (3,3))
b = np.random.randint(50, size = (3,3))
```

```python
print(a)
print(b)
```

```python
# diff = a - b
diff = np.subtract(a,b)
print(diff)
```

3) Multiplying by a scalar:

```python
m = np.random.randint(50, size = (3,3))
s = 5
product = np.multiply(m,s)
print(product)
```

4) Multiplying two matrices:

```python
a = np.random.randint(20 , size = (3,4))
b = np.random.randint(10, size = (4,2))
prod = np.dot(a,b)
```

```python
print(prod)
```

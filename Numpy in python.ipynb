**List v/s Numpy ---> time taken**

```python
from time import process_time
```

1) Time taken by a list

```python
python_list = [i for i in range(10000)] # creating a list with numbers from 0 to 9999 using a for loop
start_time = process_time()
python_list = [i+5 for i in python_list] # adding 5 to every element of the list
end_time = process_time()
print(end_time - start_time)
```

2) Time taken for numpy array

```python
np_array = np.array([i for i in range(10000)])
st = process_time()
np_array += 5
et = process_time()
print(et-st)
```

```python
list = [1,2,3,4,5]
print(list)
type(list)
```

```python
# creating a 1-d np array
np_list = np.array([1,2,3,4,5])
print(np_list)
type(np_list) # gives numpy.ndarray => ndarray is n-dimessional array, so numpy arrays are simliar to matrixes
```

#Initial place holders in numpy arrays

The initial values in the np array

```python
# a numpy array of zeros
x = np.zeros((4,6))
print(x)
```

```python
# a numpy array with all ones
y = np.ones((3,5))
print(y)
```

```python
# a numpy array with a particular value
f = np.full((3,3),7) # np.full(shape, value)
print(f)
```

```python
# creating an identity matrix
i = np.eye(5) # np.eye(size) => because in identity matrix no. of rows = no. of columns
print(i)
```

```python
# creating a np array with random values
# !!! Note that all the values will be between 0 and 1 , that are all floating point numbers
r = np.random.random((3,4))
print(r)
```

```python
# creating a random integer numpy array within a given range
ri = np.random.randint(1,100,(4,4)) # ==> np.random.randint(start,end,shape)
print(ri)
```

```python
# creating an array of evenly spaced values
evenlySpaced = np.linspace(10,30,6) # => np.linspace(start,end,no.of values we want in that range)
print(evenlySpaced)
```

```python
 # evenly spaced array with specifying the step
 e = np.arange(10,30,5) # => np.arange(start,end,step size)
 print(e)
```

**Mathematical operations that can be performed on a np_array:**

```python
l1 = [1,2,3,4]
l2 = [5,6,7]
print(l1+l2) # it concatnates the lists
```

```python
p = np.random.randint(0,10,(3,3))
q = np.random.randint(10,20,(3,3))
print(p)
print(q)
```

```python
# operations in numpy arrays , not like lists
print(p+q)
print(p-q)
print(p*q)
print(p/q)
```

```python
# another way of adding, subtracting the arrays
print(np.add(p,q))
print(np.subtract(p,q))
print(np.multiply(p,q))
print(np.divide(p,q))
```

**Array manipulation**

```python
# Transpose of a matrix
arr = np.random.randint(0,15,(2,3))
print(arr)
print(arr.shape)
```

```python
trans = np.transpose(arr)
print(trans)
print(trans.shape)
```

```python
t2 = trans.T
print(t2)
print(t2.shape)
```

```python
# Reshaping an array
s = np.random.randint(0,20,(4,3))
print(s)
print(s.shape)
```

```python
re = s.reshape(6,2)
print(re)
print(re.shape)
```

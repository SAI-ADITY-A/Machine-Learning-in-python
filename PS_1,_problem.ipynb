#1) Importing the dependencies:

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# 2) Data collection and pre-processing:

```python
breast_cancer_data = pd.read_csv('/content/data_p1.csv')
```

```python
# importing data from csv file to a pandas dataframe
breast_cancer_data.head()
```

```python
# no. of rows and columns in the data
breast_cancer_data.shape
```

```python
breast_cancer_data.info()
```

```python
# removing "Unnamed: 32" column which is unnecessary
cdata = breast_cancer_data.drop(columns = 'Unnamed: 32', axis = 1)
cancer_data = cdata.drop(columns = 'id',axis = 1)
cancer_data.head()
```

```python
#no. of rows and columns after removing the column
cancer_data.shape
#now we don't have any missing values in the dataset.
```

```python
# few important statistics of the data
cancer_data.describe()
```

```python
# checking the diagnosis for it's distribution
cancer_data['diagnosis'].value_counts()
```

 Here, B = benign(non-cancerous) and M = malignant(cancerous)

```python
# statistics of the dataset corresponding to the B and M diagnosis
cancer_data.groupby('diagnosis').mean()
```

So, from the above analysis, we can see that the radius of cells can give an insight into whether a person has cnacer of not.

For Benign : radius = 12.14

For Malignant : radius = 17.46

#3) Seperating the data (features and diagnosis):

```python
x = cancer_data.drop(columns = 'diagnosis', axis = 1)
y = cancer_data['diagnosis']
```

```python
print(x)
```

# 4) Tarinig and testing data splitting:

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

# 5) Logistic Regression Model:

```python
model = LogisticRegression()
```

```python
# feeding and training our model with the training data
model.fit(x_train, y_train)
```

# 6) Model Evaluation:

```python
x_train_prediction = model.predict(x_train)
training_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print(training_accuracy)
```

```python
print("Accuracy for the training data: ",training_accuracy)
```

Checking accuracy for test data:

```python
x_test_prediction = model.predict(x_test)
testing_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print(testing_accuracy)
```

```python
print("Accuracy for the testing data: ",testing_accuracy)
```

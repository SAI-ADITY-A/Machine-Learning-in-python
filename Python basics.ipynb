**!Note**: When we input two numbers and do "+" operation on them to sum them, it doesn't work rather it concatnates both the num1 and num2 treating them as strings. See below example:

```python
num1 = input("Enter first number: ")
num2 = input("Enter the second number: ")
sum = num1 + num2
print(sum)
```

The solution is type casting, we have to convert anything exrtacted from input function to it's respective data type, for example num1 is to be type casted into integer datatype...

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter the second number: "))
sum = num1 + num2
print(sum)
```

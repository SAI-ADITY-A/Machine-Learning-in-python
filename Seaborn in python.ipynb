#**Seaborn library**



*   Data visualization library



```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# seaborn has some built-in datasets
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head() # the dataset loaded from seaborn will be in form of pandas dataframe
```

```python
 # visualize the data
 sns.relplot(data = tips, x = 'total_bill', y = 'tip', col = 'time', hue = 'smoker', style = 'smoker', size = 'size')
```

```python
# iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.scatterplot(data = iris, x = 'sepal_length', y = 'petal_length', hue = 'species')
```

```python
sns.scatterplot(data = iris, x = 'sepal_width', y = 'petal_width', hue = 'species')
```

```python
# titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.shape
```

```python
# count plot
sns.countplot(data = titanic, x = 'class')
```

```python
# no. of people survived
sns.countplot(x = 'survived', data = titanic)
```

```python
# Bar plot
sns.barplot(x = 'sex', y = 'survived', hue = 'class', data = titanic)
```

```python
boston_df = pd.read_csv('/content/Boston.csv')
```

```python
boston_df.head()
```

```python
# Distribution plot
sns.distplot(boston_df['medv'])
```

```python
# correlation matrix
correlation = boston_df.corr()
```

```python
#Heat map
plt.figure(figsize = (10,10))
sns.heatmap(correlation, cbar = True, square = True, fmt = '.1f', annot = True, annot_kws = {'size' : 8}, cmap = 'Blues')
# cbar = True => we want the color bar
# square => we want the square shape
# fmt => no. of decimal points .2f --> 2 decimal points
# annot => the annotations, column names on the side of matrix
# annot-kws => the size of the annotations
#cmap => the color of matrix is given as Blues
```

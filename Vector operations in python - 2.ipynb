# More operations on vectors:


```python
import numpy as np
```

1) Dot product:

```python
a = np.array([2,3])
b = np.array([4,4])
a_dot_b = np.dot(a,b)
print(a_dot_b)
```

```python
c = np.array([20,34,9])
d = np.array([-9,98,-37])
c_dot_d = np.dot(c,d)
print(c_dot_d)
```

2) Cross product:

```python
a = np.array([2,3])
b = np.array([4,4])
a_cross_b = np.cross(a,b)
print(a_cross_b)
```

```python
c = np.array([10,8,7])
d = np.array([-8,7,5])
c_cross_d = np.cross(c,d)
print(c_cross_d)
```

3) Projection of one vector on another vector:

```python
a = np.array([2,5])
v = np.array([8,-6])
# magnitude of v vector
v_mag = np.sqrt(sum(v**2))
proj_a_on_v = (np.dot(a,v)/(v_mag**2))*v
```

```python
print("The projection of a vector on v vector is:", proj_a_on_v)
```

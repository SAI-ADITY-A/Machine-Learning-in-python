# Operations on Vectors in python:


1) Plotting a simple vector

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
plt.quiver(0,0,4,5)
plt.show
#this  won't the give the graph as we need, it needs to be scaled properly with some parameters
```

```python
plt.quiver(0,0,4,5,scale_units = 'xy', angles = 'xy', scale = 1)
plt.xlim(-8,8)
plt.ylim(-8,8)
```

```python
# Plotting two vectors in same plot
plt.quiver(0,0,4,5,scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,-3,4,scale_units = 'xy', angles = 'xy', scale = 1, color = 'r')
plt.quiver(0,0,-3,-7,scale_units = 'xy', angles = 'xy', scale = 1, color = 'g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show
```

2) Addition of two vectors and their plots:

```python
vector1 = np.asarray([0,0,2,3])
vector2 = np.asarray([0,0,3,-2])
sum = vector1 + vector2
print(sum)
```

```python
# Plotting these vectors
plt.quiver(0,0,2,3,scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,3,-2,scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,5,1,scale_units = 'xy', angles = 'xy', scale = 1, color = 'r')
plt.xlim(-6,6)
plt.ylim(-6,6)
plt.show
```

3) Subtraction of two vectors:

```python
vector1 = np.asarray([0,0,2,3])
vector2 = np.asarray([0,0,3,-2])
diff = vector1 - vector2
print(diff)
```

```python
# Plotting these vectors
plt.quiver(0,0,2,3,scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,3,-2,scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,-1,5,scale_units = 'xy', angles = 'xy', scale = 1, color = 'r')
plt.xlim(-6,6)
plt.ylim(-6,6)
plt.show
```

4) Multiplying a vector by a scalar:

```python
v1 = np.asarray([0,0,2,3])
v2 = 2*v1
print(v2)
```

```python
#plotting these two vectors
plt.quiver(0,0,2,3,scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,4,6,scale_units = 'xy', angles = 'xy', scale = 1,color = 'y')
plt.xlim(-6,6)
plt.ylim(-6,6)
plt.show()
```
